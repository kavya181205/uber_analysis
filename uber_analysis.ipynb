```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apr=pd.read_csv('uber-raw-data-apr14.csv')
may=pd.read_csv('uber-raw-data-may14.csv')
jun=pd.read_csv('uber-raw-data-jun14.csv')
jul=pd.read_csv('uber-raw-data-jul14.csv')
aug=pd.read_csv('uber-raw-data-aug14.csv')
sep=pd.read_csv('uber-raw-data-sep14.csv')
```

```python
df=pd.concat([apr,may,jun,jul,aug,sep])
df
```

```python
df.info()
```

```python
df=df.rename(columns={'Date/Time':'date_time'})
df
```

```python
df['date_time']=pd.to_datetime(df['date_time'])
df
```

```python
df['Month']=df['date_time'].dt.month_name()
df['Weekday']=df['date_time'].dt.day_name()
df['Day'] = df['date_time'].dt.day
df['Hour'] = df['date_time'].dt.hour
df['Minute'] = df['date_time'].dt.minute
```

```python
df_hour_grouped = df.groupby(['Hour']).count()
df_hour_grouped
# df_hour = pd.DataFrame({'Number_of_trips':df_hour_grouped.values[:,0]}, index = df_hour_grouped.index) 
# df_hour.head()
```

# Trips by Hours

```python
plt.figure(figsize=(8, 6))
plt.bar(df_hour_grouped.index, df_hour_grouped.values[:, 0],width=0.5,label='Number of trips')
plt.xticks(df_hour_grouped.index) 
plt.xlabel('Hour')
plt.ylabel('Number of trips')
plt.title('trips by hours')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left')

plt.show()
```

# Trips by Month

```python
plt.figure(figsize=(8, 6))
df_month_grouped=df.groupby(['Month']).count()
plt.bar(df_month_grouped.index,df_month_grouped.values[:,0],width=0.5,label='Number of trips')
plt.xticks(df_month_grouped.index)
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.title('Trips by Month')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left')
plt.show()
```

# Trips by WeekDay

```python
plt.figure(figsize=(8, 6))
df_Weekday_grouped=df.groupby(['Weekday']).count()
plt.bar(df_Weekday_grouped.index,df_Weekday_grouped.values[:,0],width=0.5,label='Number of trips')
plt.xticks(df_Weekday_grouped.index)
plt.xticks(rotation=45)
plt.xlabel('Weekday')
plt.ylabel('Number of Trips')
plt.title('Trips by Weekday')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper right')
plt.show()
```

# Trips by Day

```python
plt.figure(figsize=(10, 8))
df_day_grouped=df.groupby(['Day']).count()
plt.bar(df_day_grouped.index,df_day_grouped.values[:,0],width=0.5,label='Number of trips')
plt.xticks(df_day_grouped.index)
# plt.xticks(rotation=45)
plt.xlabel('Weekday')
plt.ylabel('Number of Trips')
plt.title('Trips by day')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper right')
plt.show()
```
